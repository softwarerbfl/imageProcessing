```python
import cv2
import matplotlib.pyplot as pyplot
import matplotlib.cm as cm
```

```python
#img는 원본 이미지
img = cv2.imread("../Dongkeun-OpenCV-ImgData/opencv_logo.png")
```

```python
pyplot.imshow(img) #사진에 나타나는 숫자의 단위는 픽셀
pyplot.show()
```

# split()

```python
b,g,r = cv2.split(img)
```

```python
print(b,r,g)
```

```python
#img_merge는 img를 rgb로 merge한 이미지
img_merge = cv2.merge([r,g,b])
pyplot.imshow(img_merge)
pyplot.show()
```

# Gray Scale이미지 주의할 점

`cmap="gray`" 코드 추가

```python
gray_img = cv2.imread("../Dongkeun-OpenCV-ImgData/opencv_logo.png", cv2.IMREAD_GRAYSCALE)
pyplot.imshow(gray_img, cmap="gray")
pyplot.show()
```
